## HOCKEY RINK

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc

"""Create figure"""
fig=plt.figure()
fig.set_size_inches(11, 6)
ax=fig.add_subplot(1,1,1)

"""Pitch Outline & Centre Line"""
plt.plot([-2550,2550],[1500,1500], color="black")
plt.plot([2550,-2550],[-1500,-1500], color="black")
plt.plot([-2550,-2550],[-1500,1500], color="red")
plt.plot([2550,2550],[-1500,1500], color="red")
plt.plot([0,0],[1500,-1500], color="red")

plt.plot([875,875],[1500,-1500], color="blue")
plt.plot([-875,-875],[1500,-1500], color="blue")



"""Middle point"""
centreCircle = plt.Circle((0,0),20,color="blue",fill=True)
ax.add_patch(centreCircle)
#Stredni vhazovani
centreSpot = plt.Circle((0,0),456,facecolor="none", edgecolor='blue')
ax.add_patch(centreSpot)


"""Face off"""
a1 = plt.Circle((-1900,800),20,color="red",fill=True)
ax.add_patch(a1)
#Stredni vhazovani
a11 = plt.Circle((-1900,800),480,facecolor="none", edgecolor='red')
ax.add_patch(a11)

"""Face off 2"""
a2 = plt.Circle((1900,-800),20,color="red",fill=True)
ax.add_patch(a2)

"""X CENTER"""
a22 = plt.Circle((1900,-800),480,facecolor="none", edgecolor='red')
ax.add_patch(a22)

a33 = plt.Circle((1900,800),480,facecolor="none", edgecolor='red')
ax.add_patch(a33)

"""Face off 3"""
a3 = plt.Circle((1900,800),20,color="red",fill=True)
ax.add_patch(a3)


"""Face off 4""" 
a2 = plt.Circle((-1900,-800),20,color="red",fill=True)
ax.add_patch(a2)

a22 = plt.Circle((-1900,-800),480,facecolor="none", edgecolor='red')
ax.add_patch(a22)

"""face off 5"""
a5 = plt.Circle((-750,-800),20,color="red",fill=True)
ax.add_patch(a5)

a6 = plt.Circle((750,-800),20,color="red",fill=True)
ax.add_patch(a6)

a7 = plt.Circle((-750,800),20,color="red",fill=True)
ax.add_patch(a7)

a8 = plt.Circle((750,800),20,color="red",fill=True)
ax.add_patch(a8)

#Prepare Arcs
leftArc = Arc((2550,1100),height=800,width=800,angle=0,theta1=360,theta2=100,color="black", linewidth=2)
rightArc = Arc((2550,-1100),height=800,width=800,angle=0,theta1=260,theta2=370,color="black", linewidth=2)

leftArc2 = Arc((-2550,1100),height=800,width=800,angle=0,theta1=80,theta2=185,color="black", linewidth=2)
rightArc2 = Arc((-2550,-1100),height=800,width=800,angle=0,theta1=180,theta2=270,color="black", linewidth=2)

#Draw Arcs
ax.add_patch(leftArc)
ax.add_patch(rightArc)
ax.add_patch(leftArc2)
ax.add_patch(rightArc2)

plt.plot([2950,2950],[-1200,1200], color="black")

plt.plot([-2950,-2950],[1200,-1200], color="black")

"""COLOR OF THE RINK"""
pitch="aliceblue"

hriste = plt.Rectangle((-2550,-1500), 5100,3000, color=pitch,zorder=1,alpha=0.5)
ax.add_artist(hriste)

zabrankou = plt.Rectangle((-2950,-1200), 400,2400, color=pitch,zorder=1,alpha=0.9)
ax.add_artist(zabrankou)

zabrankou1 = plt.Rectangle((2550,-1200), 400,2400, color=pitch,zorder=1,alpha=0.5)
ax.add_artist(zabrankou1)


zabrankoukololeft = Arc((2550,-1100),height=800,width=800,angle=0,theta1=260,theta2=370, color=pitch,zorder=1,alpha=0.9)
ax.add_artist(zabrankoukololeft)

#GK
GKH = Arc((-2550,0),height=350,width=350,angle=0,theta1=270,theta2=90,color="black", linewidth=1)
ax.add_patch(GKH)

GKA = Arc((2550,0),height=350,width=350,angle=0,theta1=90,theta2=270,color="black")
ax.add_patch(GKA)

porada_ref = Arc((0,-1500),height=550,width=550,angle=0,theta1=0,theta2=180,color="red")

#Corners
hl = Arc((-2550,1150),height=350,width=350,angle=0,theta1=90,theta2=180,color=pitch, linewidth=32)
ax.add_patch(hl)

al = Arc((2550,1150),height=350,width=350,angle=0,theta1=0,theta2=90,color=pitch, linewidth=32)
ax.add_patch(al)

ar = Arc((2550,-1150),height=350,width=350,angle=0,theta1=270,theta2=360,color=pitch, linewidth=32)
ax.add_patch(ar)

hr = Arc((-2550,-1150),height=350,width=350,angle=0,theta1=180,theta2=270,color=pitch, linewidth=32)
ax.add_patch(hr)

"""
x: 2950 - -2950
y: 1200 - -1200

"""
"""SHOTS"""
#shot = plt.Circle((2950/100*80,1500/100*20),50,color="green",fill=True)
#ax.add_patch(shot)

plt.show()
```
