### Create dictionary

```python
a = {"Jersey number":[1, 2, 3, 4, 5, 6], 
     "Name":["Luke", "Josh", "David", "Joe", "John", "Adam"],
     "Surname":["Blake", "Smith", "Brown", "Garcia", "Davis", "Lee"],
    "Team":["Shark", "Shark", "Shark", "Shark", "Shark", "Shark"],
    "Age":[25, 22, 27, 35, 31, 20]}
```

### Create dataframe from dictionary

```python
b = pd.DataFrame(a)
b
```

```python
#Create list
c =  [[7, "Eric", "Adans", "Shark", 33], [8, "Wayne", "Lambert", "Shark", 37 ]]

#Create dataframe from list
d = pd.DataFrame(c)
print(d)
print()
# Names of columns
print("Names of columns are: " + str(d.columns))
```

### Rename columns

```python
d.columns = "Jersey number", "Name", "Surname", "Team", "Age"
d
```

### Append table D on table B

```python
players = b.append(d)
players
```

### Reset index + remember for next operations

```python
players.reset_index(inplace = True)
```

### Delete column index + keep this for next operations => inplace=True

```python
players.drop('index', 1, inplace = True)
```

```python
#Create dictionary
e = {"Jersey number":[1, 2, 3, 4, 5, 6, 7, 8], 
    "Matches":[52, 48, 40, 52, 52, 35, 45, 20],
    "Goals":[20, 15, 12, 5, 30, 10, 7, 5],
    "Asists":[22, 30, 25, 7, 20, 11, 10, 8]    
    }

#Create dataframe
points = pd.DataFrame(e)
points
```

### Merge dataframes

```python
teamshark = players.merge(points, left_on="Jersey number", right_on="Jersey number", how="inner")
teamshark
```

### Add columns

```python
teamshark["Points"] = teamshark["Goals"] + teamshark["Asists"]

teamshark["G per match"] = round(teamshark["Goals"] / teamshark["Matches"], 2)
teamshark["A per match"] = round(teamshark["Asists"] / teamshark["Matches"], 2)
teamshark["P per match"] = round(teamshark["Points"] / teamshark["Matches"], 2)
teamshark
```
